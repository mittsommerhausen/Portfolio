# Предобработка

```python
import pandas as pd
import numpy as np
```

```python
data = r'C:\Users\polin\Downloads\ba_first.xlsx'
xl = pd.ExcelFile(data)
xl.sheet_names
```

```python
y_2019 = xl.parse('2019')
y_2020 = xl.parse('2020')
display(y_2019)
y_2020
```

```python
y_2019.rename(columns={y_2019.columns[0]: 'order_number', y_2019.columns[1]: 'store_number', y_2019.columns[2]: 'item_id', y_2019.columns[3]: 'date', y_2019.columns[4]: 'total'},inplace=True)
```

```python
y_2020.rename(columns={y_2020.columns[0]: 'order_number', y_2020.columns[1]: 'store_number', y_2020.columns[2]: 'item_id', y_2020.columns[3]: 'date', y_2020.columns[4]: 'total'},inplace=True)
```

```python
display(y_2019.info(), y_2020.info())
```

Как мы видим, колоссальная часть данных отсутствует. В рабочем проекте необходимо было бы запросить причину отсутствия данных у заказчика, однако сейчас просто избавимся от пропусков, чтобы они не мешали в работе:

```python
y_2019 = y_2019.dropna()
y_2020 = y_2020.dropna()
display(y_2019.info(), y_2020.info())
```

```python
to_int = ['order_number', 'store_number', 'item_id']
for val in to_int:
    y_2020[val] = y_2020[val].astype('int')
    y_2019[val] = y_2019[val].astype('int')
```

# Аналитика

```python
y19 = list(y_2019['item_id'].unique())
y20 = list(y_2020['item_id'].unique())
y20u = []
for i in y20:
    if i not in y19:
        y20u.append(i)
y19o = []
for i in y20:
    if i in y19:
        y19o.append(i)
```

```python
y20u = list(set(y20u))
y19, y20u
```

Сформируем таблицы с новыми и старыми артикулами:

```python
new_20 = y_2020.query('item_id in @y20u')
new_20.head(10)
```

```python
old_20 = y_2020.query('item_id in @y19')
old_20.head(10)
```

```python
new_20['status'] = 'NEW'
old_20['status'] = 'OLD'
display(new_20.head(10))
old_20.head(10)
```

```python
full_20 = pd.concat([new_20, old_20], axis=0)
full_20 = full_20.sort_index(ascending=True)
full_20.head(15)
```

```python
full_20.to_excel(r'C:\Users\polin\Downloads\business_analysis_2019_2020.xlsx', sheet_name='OLD_NEW', index=False)
```

```python
old_19 = y_2019.query('item_id in @y19o')
old_19.head(10)
```

```python
old_19_sum = old_19.pivot_table(index='item_id', values='total', aggfunc='sum').reset_index()
old_20_sum = old_20.pivot_table(index='item_id', values='total', aggfunc='sum').reset_index()
```

```python
old_total = pd.concat([old_19_sum, old_20_sum], axis=1)
old_total.columns.values[0] = 'item_id_19'
old_total.columns.values[1] = 'sum_total_19'
old_total.columns.values[2] = 'item_id_20'
old_total.columns.values[3] = 'sum_total_20'
old_total
```

```python
old_total.to_excel(r'C:\Users\polin\Downloads\business_analysis_2019_2020.xlsx', sheet_name='OLD_TOTAL', index=False)
```
